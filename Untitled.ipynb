### 1.Дерева рішень

Блок-схема — це деревоподібна структура запитань так/ні, яка веде вас через певний процес на основі вашої конкретної ситуації. Дерево рішень, по суті, є блок-схемою для прийняття рішення, як класифікувати спостереження: воно складається з серії рішень так/ні або якщо/інше, які в кінцевому підсумку приписують кожне спостереження певній ймовірності або класу. Серію рішень так/ні можна зобразити як серію гілок, які ведуть рішення або «листя» у нижній частині дерева.


```python
import numpy as np
import pandas as pd
```

```python
# Load and prepare Titanic data
titanic_train = pd.read_csv("input/train.csv")    # Read the data

# Impute median Age for NA Age values
new_age_var = np.where(titanic_train["Age"].isnull(), # Logical check
                       28,                       # Value if check is true
                       titanic_train["Age"])     # Value if check is false

titanic_train["Age"] = new_age_var 
```

Далі потрібно завантажити та ініціалізувати модель дерева рішень scikit-learn, а потім навчити модель за допомогою змінної sex:

```python
from sklearn import tree
from sklearn import preprocessing
```

```python
# Initialize label encoder
label_encoder = preprocessing.LabelEncoder()

# Convert Sex variable to numeric
encoded_sex = label_encoder.fit_transform(titanic_train["Sex"])

# Initialize model
tree_model = tree.DecisionTreeClassifier()

# Train the model
tree_model.fit(X = pd.DataFrame(encoded_sex), 
               y = titanic_train["Survived"])
```

Зверніть увагу на список аргументів за замовчуванням, включених до моделі вище. Детальніше про них читайте тут.

Тепер давайте розглянемо візуалізацію дерева, створеного моделлю. Ми можемо зробити це за допомогою бібліотеки "graphviz":

```python
import graphviz

# Save tree as dot file
dot_data = tree.export_graphviz(tree_model, out_file=None) 
graph = graphviz.Source(dot_data)  
graph 
```

Графік дерева показує нам, що він складається лише з одного вузла прийняття рішень, який розбиває дані про змінну стать (перша змінна, X[0]). Усі 314 жінок потрапляють в один листовий вузол, а всі 577 чоловіків – в інший листовий вузол.

Давайте зробимо прогнози з цим деревом і переглянемо таблицю результатів:

```python
# Get survival probability
preds = tree_model.predict_proba(X = pd.DataFrame(encoded_sex))

pd.crosstab(preds[:,0], titanic_train["Sex"])
```

Таблиця показує, що дереву рішень вдалося створити просту гендерну модель, де всі жінки виживають, а всі чоловіки гинуть.

Давайте створимо нове дерево рішень, яке додасть змінну класу пасажира, і подивимося, як вона змінить результати прогнозів:

```python
# Make data frame of predictors
predictors = pd.DataFrame([encoded_sex, titanic_train["Pclass"]]).T

# Train the model
tree_model.fit(X = predictors, 
               y = titanic_train["Survived"])
```

Тепер давайте подивимося на графік нової моделі дерева рішень:

```python
# Save tree as dot file
dot_data = tree.export_graphviz(tree_model, out_file=None) 
graph = graphviz.Source(dot_data)  
graph 
```

Зверніть увагу, що якщо додати ще одну змінну, дерево буде значно складнішим. Тепер він має 6 вузлів рішень, 6 вузлів листків і максимальну глибину 3. Давайте зробимо прогнози та переглянемо таблицю результатів:

```python
# Get survival probability
preds = tree_model.predict_proba(X = predictors)

# Create a table of predictions by sex and class
pd.crosstab(preds[:,0], columns = [titanic_train["Pclass"], 
                                   titanic_train["Sex"]])
```

Зауважте, що більш складна модель все ще передбачає вищий рівень виживання для жінок, ніж для чоловіків, але жінки третього класу мають лише 50% передбачуваної ймовірності смерті, тоді як жінки першого класу, за прогнозами, помруть менше ніж у 5% випадків.

Чим більше змінних ви додаєте до дерева рішень, тим більше рішень «так/ні» воно може прийняти, що призводить до глибшого та складнішого дерева. Однак надмірна складність дерева рішень робить його схильним до переобладнання навчальних даних, що може призвести до поганого узагальнення невидимих даних. Давайте дослідимо, створивши більше дерево з кількома змінними:

```python
predictors = pd.DataFrame([encoded_sex,
                           titanic_train["Pclass"],
                           titanic_train["Age"],
                           titanic_train["Fare"]]).T

# Initialize model with maximum tree depth set to 8
tree_model = tree.DecisionTreeClassifier(max_depth = 8)

tree_model.fit(X = predictors, 
               y = titanic_train["Survived"])
```

```python
# Save tree as dot file
dot_data = tree.export_graphviz(tree_model, out_file=None) 
graph = graphviz.Source(dot_data)  
graph 
```

Зображення вище ілюструє, наскільки складними можуть стати дерева рішень, коли ви почнете додавати більше пояснювальних змінних. Ви можете контролювати складність дерева, змінюючи деякі параметри функції дерева рішень за замовчуванням. Наприклад, коли ми створили дерево вище, ми встановили max_depth = 8, що обмежило дерево глибиною 8 (якби ми цього не зробили, дерево було б набагато більше!).

Заради інтересу, давайте перевіримо точність цієї моделі дерева рішень на даних для навчання:

```python
tree_model.score(X = predictors, 
                 y = titanic_train["Survived"])
```

Модель майже на 89% точна щодо даних для навчання, але як вона діє на невидимих даних? Давайте завантажимо тестові дані, зробимо деякі прогнози, надішліть їх до Kaggle, щоб дізнатися:

```python
# Read and prepare test data
titanic_test = pd.read_csv("input/test.csv")    # Read the data

# Impute median Age for NA Age values
new_age_var = np.where(titanic_test["Age"].isnull(), # Logical check
                       28,                       # Value if check is true
                       titanic_test["Age"])      # Value if check is false

new_fare_var = np.where(titanic_test["Fare"].isnull(), # Logical check
                       50,                       # Value if check is true
                       titanic_test["Fare"])      # Value if check is false

titanic_test["Age"] = new_age_var 
titanic_test["Fare"] = new_fare_var
```

```python
# Convert test variables to match model features
encoded_sex_test = label_encoder.fit_transform(titanic_test["Sex"])

test_features = pd.DataFrame([encoded_sex_test,
                              titanic_test["Pclass"],
                              titanic_test["Age"],
                              titanic_test["Fare"]]).T
```

Після подання модель оцінює точність 0,78468, що трохи краще, ніж проста модель на основі статі, але набагато гірше, ніж точність, яку модель досягла на самих навчальних даних. Це підкреслює той факт, що прогнозна ефективність на навчальних даних є поганим барометром ефективності прогнозування нових даних.

## Перевірка очікування та перехресна перевірка

Створюючи прогнозну модель, ми хотіли б отримати точне уявлення про її здатність узагальнювати невидимі дані, перш ніж фактично використовувати її для невидимих ​​даних. Як ми бачили раніше, створення прогнозів на самих навчальних даних для перевірки точності моделі працює не дуже добре: складна модель може дуже точно відповідати навчальним даним, але не узагальнюватися на нові, невидимі дані. Ми можемо краще зрозуміти очікувану продуктивність моделі на невидимих ​​даних, відклавши частину наших навчальних даних під час створення моделі, а потім використавши ці відкладені дані для оцінки ефективності моделі. Ця техніка відкладення деяких навчальних даних для оцінки здатності моделі до узагальнення відома як перевірка.

Перевірка очікування та перехресна перевірка є двома поширеними методами оцінки моделі перед її використанням на тестових даних. Перевірка очікування передбачає поділ навчальних даних на дві частини, навчальний набір і набір перевірки, побудову моделі з навчальним набором, а потім оцінку ефективності за допомогою набору перевірки. Теоретично, продуктивність моделі на наборі підтвердженої перевірки має приблизно відображати продуктивність, яку ви очікуєте побачити на невидимих ​​тестових даних. На практиці перевірка затримки швидка і може добре працювати, особливо на великих наборах даних, але вона має деякі підводні камені.

Резервування частини навчальних даних для набору затримки означає, що ви не використовуєте всі дані, які є у вашому розпорядженні, для побудови моделі на етапі перевірки. Це може призвести до неоптимальної продуктивності, особливо в ситуаціях, коли ви не маєте багато даних для роботи. Крім того, якщо ви використовуєте той самий набір перевірки затримки для оцінки занадто великої кількості різних моделей, ви можете в кінцевому підсумку знайти модель, яка добре відповідає набору перевірки через випадковість, яка не обов’язково добре узагальнюватиметься на невидимі дані. Незважаючи на ці недоліки, варто навчитися використовувати набір перевірки затримки в Python.

Ви можете створити набір перевірки затримки за допомогою train_test_split() у бібліотеці cross_validation sklearn:

```python
from sklearn.model_selection import train_test_split
```

```python
v_train, v_test = train_test_split(titanic_train,     # Data set to split
                                   test_size = 0.25,  # Split ratio
                                   random_state=1,    # Set random seed
                                   stratify = titanic_train["Survived"]) #*

# Training set size for validation
print(v_train.shape)
# Test set size for validation
print(v_test.shape)
```

Примітка: під час виконання класифікації бажано, щоб кожен клас у цільовій змінній мав приблизно однакову пропорцію для кожного розділення даних. Аргумент стратифікації дозволяє вказати цільову змінну для рівномірного розподілу між потягом і тестом.

Результати вище показують, що ми успішно створили новий навчальний набір з приблизно 75% вихідних даних і тестовий набір для перевірки з 25% даних. Ми могли б продовжити, створюючи моделі з цим новим навчальним набором і роблячи прогнози щодо набору перевірки, щоб оцінити моделі.

Перехресна перевірка є популярною альтернативою перевірці, яка передбачає поділ навчальних даних на два або більше розділів і створення моделі для кожного розділу, де розділ діє як набір перевірки, а решта даних — як навчальні дані. Поширеною формою перехресної перевірки є перехресна перевірка "k-fold", яка випадковим чином розбиває дані на деяку кількість k (параметр, визначений користувачем) розділів, а потім створює k моделей, кожна з яких перевіряється на одному з розділів. Кожна з k моделей потім об’єднується в одну підсумкову модель.

Основна перевага перехресної перевірки полягає в тому, що вона використовує всі навчальні дані для побудови та оцінки остаточної моделі. Основним недоліком є ​​те, що створення та тестування кількох моделей може бути дорогим у обчислювальному відношенні, тому воно, як правило, займає набагато більше часу, ніж перевірка, що зупиняється. Ви можете створити K перехресної перевірки даних, використовуючи функцію Kfold() у бібліотеці model_selection sklearn:

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=10)
kf.get_n_splits(titanic_train)
```

Після створення об’єкта перехресної перевірки ви можете перейти до кожного згину та тренувати та оцінювати свою модель для кожного з них:

```python
fold_accuracy = []

titanic_train["Sex"] = encoded_sex

for train_fold, valid_fold in kf.split(titanic_train):
    train = titanic_train.loc[train_fold] # Extract train data with cv indices
    valid = titanic_train.loc[valid_fold] # Extract valid data with cv indices
    
    model = tree_model.fit(X = train[["Sex","Pclass","Age","Fare"]], 
                           y = train["Survived"])
    valid_acc = model.score(X = valid[["Sex","Pclass","Age","Fare"]], 
                            y = valid["Survived"])
    fold_accuracy.append(valid_acc)    

print("Accuracy per fold: ", fold_accuracy, "\n")
print("Average accuracy: ", sum(fold_accuracy)/len(fold_accuracy))
```

Точність моделі може значно відрізнятися від однієї частини до іншої, особливо для невеликих наборів даних, але середня точність у всіх згинах дає вам уявлення про те, як модель може працювати з невидимими даними.

Як і під час перевірки затримки, ми хотіли б, щоб класи цільової змінної мали приблизно однакову пропорцію в кожній частині під час виконання перехресної перевірки для проблеми класифікації. Щоб виконати багатошарову перехресну перевірку, використовуйте функцію StratifiedKFold() замість KFold().

Ви можете використовувати оцінку моделі зі стратифікованою перехресною перевіркою за допомогою одного виклику функції за допомогою функції cross_val_score():

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(estimator= tree_model,     # Model to test
                X= titanic_train[["Sex","Pclass",   # Train Data
                                  "Age","Fare"]],  
                y = titanic_train["Survived"],      # Target variable
                scoring = "accuracy",               # Scoring metric    
                cv=10)                              # Cross validation folds

print("Accuracy per fold: ")
print(scores)
print("Average accuracy: ", scores.mean())
```

Зверніть увагу, що середня точність у кожній складці вища, ніж у прикладі K-складок без стратифікації. Функція cross_val_score корисна для тестування моделей та налаштування параметрів моделі (знаходження оптимальних значень для таких аргументів, як максимальна глибина дерева, що впливають на продуктивність моделі).

## 🌳 Дерево рішень і випадковий ліс

## 1. Дерево рішень

Дерева рішень є важливим типом алгоритму для прогнозного моделювання машинного навчання.

Класичні алгоритми дерева рішень існують протягом десятиліть, а сучасні варіанти, як-от випадковий ліс, є одними з найпотужніших доступних методів.

Дерева класифікації та регресії або скорочено CART – це термін, введений Лео Брейманом для позначення алгоритмів дерева рішень, які можна використовувати для задач моделювання з класифікацією або прогнозуванням регресії.

Класично цей алгоритм називають «деревами рішень», але на деяких платформах, таких як R, вони називаються більш сучасним терміном CART.

Алгоритм CART забезпечує основу для таких важливих алгоритмів, як дерева рішень у пакетах, випадкові ліси та дерева рішень, що розширюються.
### Представлення моделі CART
Представленням моделі CART є бінарне дерево.

Це ваше двійкове дерево з алгоритмів і структур даних, нічого зайвого. Кожен кореневий вузол представляє одну вхідну змінну (x) і точку розділення цієї змінної (припускаючи, що змінна є числовою).

Листові вузли дерева містять вихідну змінну (y), яка використовується для передбачення.

З огляду на новий вхід, дерево обходить шляхом оцінки конкретного входу, розпочатого в кореневому вузлі дерева.

#### Деякі переваги дерев рішень:
* Прості для розуміння та інтерпретації.
* Дерева можна візуалізувати.
* Вимагають невеликої підготовки даних.
* Здатні обробляти як числові, так і категоріальні дані.
* Можливість перевірити модель за допомогою статистичних тестів.
* Працює добре, навіть якщо його припущення дещо порушуються справжньою моделлю, на основі якої були згенеровані дані.

#### До недоліків дерев рішень можна віднести:
* Переобладнання. Щоб уникнути цієї проблеми, необхідні такі механізми, як обрізка (наразі не підтримується), встановлення мінімальної кількості зразків, необхідної для листового вузла, або встановлення максимальної глибини дерева.
* Дерева рішень можуть бути нестабільними. Пом’якшення: використовуйте дерева рішень в ансамблі.
* Не може гарантувати повернення глобально оптимального дерева рішень. Пом’якшення: навчання кількох дерев у учня ансамблю
* Учні дерева рішень створюють упереджені дерева, якщо деякі класи домінують. Рекомендація: збалансуйте набір даних перед встановленням

```python
df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis="columns", inplace=True)

categorical_col = []
for column in df.columns:
    if df[column].dtype == object and len(df[column].unique()) <= 50:
        categorical_col.append(column)
        
df['Attrition'] = df.Attrition.astype("category").cat.codes
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('Attrition', axis=1)
y = df.Attrition

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Застосування алгоритмів дерева та випадкового лісу

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        clf_report = pd.DataFrame(classification_report(y_train, pred, output_dict=True))
        print("Train Result:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, pred)}\n")
        
    elif train==False:
        pred = clf.predict(X_test)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))
        print("Test Result:\n================================================")        
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test, pred)}\n")
```

## 1. Класифікатор дерева рішень
Параметри дерева рішень:

- `criterion`: функція вимірювання якості розщеплення. Підтримуваними критеріями є "`gini`" для домішки Джині та "`entropy`" для отримання інформації.
***
- `splitter`: Стратегія, яка використовується для вибору розщеплення в кожному вузлі. Підтримувані стратегії: "`best`" для вибору найкращого розподілу та "`random`" для вибору найкращого випадкового розподілу.
***
- `max_depth`: Максимальна глибина дерева. Якщо немає, то вузли розгортаються до тих пір, поки всі листи не стануть чистими або доки всі листи не містять менше ніж `min_samples_split` вибірок.
***
- `min_samples_split`: Мінімальна кількість вибірок, необхідних для розділення внутрішнього вузла.
***
- `min_samples_leaf`: Мінімальна кількість зразків, яка повинна бути у вузлі листка. Точка розщеплення на будь-якій глибині буде розглядатися, лише якщо вона залишає принаймні навчальні вибірки ``min_samples_leaf`` у кожній з лівої та правої гілок. Це може мати ефект згладжування моделі, особливо в регресії.
***
- `min_weight_fraction_leaf`: Мінімальна зважена частка від загальної суми ваг (всіх вхідних вибірок), необхідна для лівого вузла. Зразки мають однакову вагу, якщо sample_weight не вказано.
***
- `max_features`: Кількість функцій, які слід враховувати при пошуку найкращого розділення.
***
- `max_leaf_nodes`: Вирощує дерево з ``max_leaf_nodes`` найкращим способом. Найкращі вузли визначаються як відносне зменшення домішок. Якщо немає, то необмежену кількість листових вузлів.
***
- `min_impurity_decrease`: Вузол буде розщепленим, якщо це розщеплення спричинить зменшення домішки більше або дорівнює цьому значенню.
***
- `min_impurity_split`: Поріг для ранньої зупинки росту дерев. Вузол розділиться, якщо його домішка буде вище порога, інакше це листок.

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)

print_score(tree_clf, X_train, y_train, X_test, y_test, train=True)
print_score(tree_clf, X_train, y_train, X_test, y_test, train=False)
```

## Знайомство з алгоритмом дерева рішень

Алгоритм дерева рішень є одним з найпопулярніших алгоритмів машинного навчання. Він використовує деревоподібну структуру та їх можливі комбінації для вирішення певної проблеми. Він належить до класу алгоритмів навчання з керівництвом, де його можна використовувати як для цілей класифікації, так і для регресії.

Дерево рішень — це структура, яка включає кореневий вузол, гілки та листкові вузли. Кожен внутрішній вузол позначає тест за атрибутом, кожна гілка позначає результат тесту, а кожен листовий вузол містить мітку класу. Найвищим вузлом дерева є кореневий вузол.

Ми робимо деякі припущення під час реалізації алгоритму дерева рішень. Вони перераховані нижче:

1. На початку весь навчальний набір розглядається як корінь.
2. Значення ознак мають бути категоричними. Якщо значення неперервні, то перед побудовою моделі їх дискретизують.
3. Записи розподіляються рекурсивно на основі значень атрибутів.
4. Порядок розміщення атрибутів як кореня або внутрішнього вузла дерева здійснюється за допомогою певного статистичного підходу.

Я опишу термінологію дерева рішень у наступному розділі.

## Дерева класифікації та регресії (CART)

Нині алгоритм дерева рішень відомий під своєю сучасною назвою CART, що означає дерева класифікації та регресії. Дерева класифікації та регресії або CART – це термін, введений Лео Брейманом для позначення алгоритмів дерева рішень, які можна використовувати для задач класифікації та регресійного моделювання.

Алгоритм CART забезпечує основу для інших важливих алгоритмів, таких як дерева рішень у пакетах, випадкові ліси та дерева рішень з прискореним процесом. У цьому ядрі я вирішу проблему класифікації. Отже, я буду називати алгоритм також задачею класифікації дерева рішень.

## Термінологія алгоритму дерева рішень

* В алгоритмі «Дерево рішень» існує деревоподібна структура, в якій кожен внутрішній вузол представляє тест за атрибутом, кожна гілка представляє результат тесту, а кожен листовий вузол представляє мітку класу. Шляхи від кореневого вузла до листового вузла представляють правила класифікації.

* Ми бачимо, що в алгоритмі дерева рішень є певна термінологія. У алгоритмі дерева рішень використовуються такі терміни:

## Кореневий вузол

* Він представляє всю сукупність або вибірку. Це далі поділяється на два або більше однорідних наборів.

## Розщеплення

* Це процес поділу вузла на два або більше підвузлів.

## Вузол прийняття рішень

* Коли підвузол розбивається на інші підвузли, він називається вузлом прийняття рішень.

## Листок/кінцевий вузол

* Вузли, які не розщеплюються, називаються листовими або кінцевими вузлами.

## Обрізка

* When we remove sub-nodes of a decision node, this process is called pruning. It is the opposite process of splitting.

## Гілка/піддерево

* Підсекція цілого дерева називається гілкою або піддеревом.

## Батьківський і дочірній вузол

* Вузол, розділений на підвузли, називається батьківським вузлом підвузлів, де підвузли є дочірніми вузлами батьківського вузла.

## Інтуїція алгоритму дерева рішень

Алгоритм «Дерево рішень» є одним із найбільш поширених і широко використовуваних алгоритмів машинного навчання з наглядом, який можна використовувати як для завдань класифікації, так і для задач регресії. Інтуїція, яка лежить в основі алгоритму дерева рішень, дуже проста для розуміння.

Інтуїція алгоритму дерева рішень виглядає наступним чином:

1. Для кожного атрибута в наборі даних алгоритм дерева рішень формує вузол. Найважливіший атрибут розміщується в кореневому вузлі.

2. Щоб оцінити поставлене завдання, ми починаємо з кореневого вузла і йдемо вниз по дереву, дотримуючись відповідного вузла, який відповідає нашій умові або рішенню.

3. Цей процес триває до тих пір, поки не буде досягнутий листковий вузол. Він містить передбачення або результат дерева рішень.

## Заходи відбору атрибутів

Основною проблемою в реалізації дерева рішень є визначення атрибутів, які ми розглядаємо як кореневий вузол і кожен рівень. Цей процес відомий як вибір атрибутів. Існують різні міри вибору атрибутів, щоб ідентифікувати атрибут, який можна вважати кореневим вузлом на кожному рівні.

Існує 2 популярних заходи вибору атрибутів. Вони такі:

* Отримання інформації

* Індекс Джині

Використовуючи приріст інформації як критерій, ми припускаємо, що атрибути є категоричними, а для індексу Джіні атрибути вважаються безперервними. Ці заходи вибору атрибутів описані нижче.

## Отримання інформації

Використовуючи як критерій отримання інформації, ми намагаємося оцінити інформацію, що міститься в кожному атрибуті. Щоб зрозуміти концепцію отримання інформації, нам потрібно знати інше поняття, яке називається ентропією.

## Ентропія

Ентропія вимірює домішку в даному наборі даних. У фізиці та математиці ентропія називається випадковістю або невизначеністю випадкової величини X. У теорії інформації це відноситься до домішки в групі прикладів. Підвищення інформації – це зменшення ентропії. Приріст інформації обчислює різницю між ентропією до поділу та середньою ентропією після поділу набору даних на основі заданих значень атрибутів.

Ентропія представлена такою формулою:

![Entropy](http://www.learnbymarketing.com/wp-content/uploads/2016/02/entropy-formula.png)

Тут **c** – кількість класів, а **pi** – ймовірність, пов’язана з i-им класом.

Алгоритм дерева рішень ID3 (ітеративний дихотомайзер) використовує ентропію для обчислення приросту інформації. Отже, обчислюючи зменшення міри ентропії кожного атрибута, ми можемо розрахувати їх інформаційний приріст. Як атрибут розщеплення у вузлі вибирається атрибут з найбільшим посиленням інформації.

## Індекс Джині

Іншим показником вибору атрибутів, який використовує CART (Дерева категорій і регресії), є індекс Джіні. Він використовує метод Джіні для створення точок розщеплення.

Індекс Джіні можна представити наступною схемою:

![Gini index](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRzYHkcmZKKp2sJN1HpHvw-NgqbD9EnapnbXozXRgajrSGvEnYy&s)

Тут знову **c** – кількість класів, а **pi** – ймовірність, пов’язана з i-им класом.

Індекс Джіні говорить, що якщо ми випадковим чином вибираємо два елементи з сукупності, вони повинні бути одного класу, і ймовірність цього дорівнює 1, якщо сукупність є чистою.

Він працює з категоріальною цільовою змінною «Успіх» або «Невдача». Він виконує лише двійкові розбиття. Чим вище значення Джіні, тим вище однорідність. CART (дерево класифікації та регресії) використовує метод Джіні для створення двійкових розщеплень.

Кроки для розрахунку Джіні для розколу

1. Обчисліть Джіні для підвузлів, використовуючи формулу суми квадрата ймовірності успіху та невдачі (p^2+q^2).

2. Обчисліть Джіні для розщеплення, використовуючи зважений бал Джіні для кожного вузла цього розщеплення.

У випадку атрибута з дискретним значенням, підмножина, яка дає мінімальний індекс Джіні для цього вибраного, вибирається як атрибут розщеплення. У випадку атрибутів з безперервними значеннями стратегія полягає у виборі кожної пари сусідніх значень як можливої точки розщеплення та точки з меншим індексом Джіні, обраної як точка розщеплення. Атрибут з мінімальним індексом Джіні вибирається як атрибут розщеплення.

## Переобладнання в алгоритмі дерева рішень

Переобладнання є практичною проблемою під час побудови моделі дерева рішень. Проблема переобладнання розглядається, коли алгоритм продовжує йти все глибше і глибше, щоб зменшити помилку навчального набору, але призводить до збільшення помилки тестового набору. Таким чином, точність прогнозу для нашої моделі знижується. Зазвичай це трапляється, коли ми створюємо багато гілок через відхилення та порушення в даних.

Щоб уникнути переобладнання, можна використовувати два підходи:

* Попередня обрізка

* Після обрізка

## Попередня обрізка

Під час попередньої обрізки будівництво дерева припиняємо трохи раніше. Ми вважаємо за краще не розбивати вузол, якщо його показник якості нижче порогового значення. Але вибрати відповідну точку зупинки важко.

## Після обрізка

Під час післяобрізки ми все глибше заглиблюємося в дерево, щоб побудувати повне дерево. Якщо на дереві виявляється проблема переобрізання, обрізка виконується як етап після обрізки. Ми використовуємо дані перехресної перевірки, щоб перевірити ефект нашого обрізання. Використовуючи дані перехресної перевірки, ми перевіряємо, чи призведе розширення вузла до покращення чи ні. Якщо це показує покращення, ми можемо продовжити, розширивши цей вузол. Але якщо він показує зниження точності, його не слід розширювати. Отже, вузол слід перетворити на листовий вузол.

## Матриця плутанини

Матриця плутанини — це інструмент для підсумовування ефективності алгоритму класифікації. Матриця плутанини дасть нам чітке уявлення про продуктивність моделі класифікації та типи помилок, які створює модель. Він дає нам підсумок правильних і неправильних прогнозів, розбитих за кожною категорією. Резюме представлено у вигляді таблиці.

Під час оцінки ефективності моделі класифікації можливі чотири типи результатів. Ці чотири результати описані нижче:

**True Positives (TP)** – істинні позитиви виникають, коли ми передбачаємо, що спостереження належить до певного класу, а спостереження насправді належить до цього класу.

**True Negatives (TN)** – істинні негативи виникають, коли ми прогнозуємо, що спостереження не належить до певного класу, а спостереження фактично не належить до цього класу.

**False Positives (FP)** – хибні позитиви виникають, коли ми передбачаємо, що спостереження належить до певного класу, але фактично спостереження не належить до цього класу. Цей тип помилки називається помилкою типу I.

**False Negatives (FN)** – хибні негативи виникають, коли ми передбачаємо, що спостереження не належить до певного класу, але спостереження насправді належить до цього класу. Це дуже серйозна помилка, і вона називається помилкою типу II.

Ці чотири результати підсумовані в матриці плутанини, наведеній нижче.

```python
# Print the Confusion Matrix and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_en)

print('Confusion matrix\n\n', cm)
```
